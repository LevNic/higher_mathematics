#### 3. Задание (в программе)

Напишите код на Python, реализующий построение графиков:


```python
# Окружности
import numpy as np
import matplotlib.pyplot as plt
from pylab import rcParams
rcParams['figure.figsize'] = 4, 4


x = []
y1 = []
y2 = []
r = 1
for i in range(-100, 101):
    xx = i/100
    x.append(xx)
    y1.append(np.sqrt(r**2-xx**2))
    y2.append(-np.sqrt(r**2-xx**2))
    
plt.plot(x, y1)
plt.plot(x, y2)
plt.xlabel('X')
plt.ylabel('Y')
```

```python
# эллипса
x = []
y1 = []
y2 = []
a = 3
b = 2
for i in range(-300, 301):
    xx = i/100
    x.append(xx)
    y1.append(np.sqrt((1-xx**2/a**2)*b**2))
    y2.append(-np.sqrt((1-xx**2/a**2)*b**2))
    
plt.axis([-3.5, 3.5, -3.5, 3.5])
plt.plot(x, y1)
plt.plot(x, y2)
plt.xlabel('X')
plt.ylabel('Y')
```

```python
# гиперболы
y = []
x1 = []
x2 = []
a = 1
b = 1
for i in range(-300, 301):
    yy = i/100
    #if xx <= -a or xx >= a:
    y.append(yy)
    x1.append(np.sqrt((1+yy**2/b**2)/a**2))
    x2.append(-np.sqrt((1+yy**2/b**2)/a**2))
    
plt.axis([-3.5, 3.5, -3.5, 3.5])
plt.plot(x1, y)
plt.plot(x2, y)
plt.xlabel('X')
plt.ylabel('Y')
```

#### 5. Задание (в программе)

1. Нарисуйте трехмерный график двух параллельных плоскостей.

```python
%matplotlib inline
import matplotlib.pyplot as plt
from pylab import *
from mpl_toolkits.mplot3d import Axes3D
```

```python
fig = figure()
ax = Axes3D(fig)
X = np.arange(-5, 5, 0.5)
Y = np.arange(-5, 5, 0.5)
X, Y = np.meshgrid(X, Y)
Z = 10*X + 2*Y
Z1 = 10*X + 2*Y + 100
ax.plot_surface(X, Y, Z)
ax.plot_surface(X, Y, Z1)
show()
```

2. Нарисуйте трехмерный график двух любых поверхностей второго порядка.

```python
fig = figure()
ax = Axes3D(fig)
X = np.arange(-5, 5, 0.5)
Y = np.arange(-5, 5, 0.5)
X, Y = np.meshgrid(X, Y)
Z = X**2 + Y**2
ax.plot_surface(X, Y, Z)
ax.plot_surface(X, Y, Z + 50)
show()
```
