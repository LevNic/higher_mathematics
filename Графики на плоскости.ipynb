Нарисуйте график функции:  
y(x) = k∙cos(x – a) + b  
для некоторых (2-3 различных) значений параметров k, a, b

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(-2*np.pi, 2*np.pi, 201)
k = [1, 2, 3]
a = [ -1, 0, 1]
b = [-2, 2, 4,]
plt.grid()
for i in range(len(k)):
    y = k[i]*np.cos(x - a[i]) + b[i]
    plt.plot(x, y)
```

1. Решите систему уравнений:

```python
from scipy.optimize import fsolve

def equations(p):
    x, y = p
    return (y - x**2 + 1, np.exp(x) - x*(1 - y) - 1)

x1, y1 = fsolve(equations, (1, 1))
x2, y2 = fsolve(equations, (-1, -1))

print(x1, y1)
print(x2, y2)
```
